```python
data = pd.read_csv("/Users/Dorene/Desktop/model-dev-coding-exercise-public-master 2/plantParameters/Plant_Parameters.csv")
Henry_Hub = pd.read_csv("/Users/Dorene/Desktop/model-dev-coding-exercise-public-master 2/fuelPrices/Henry Hub.csv")
GDA_TETSTX = pd.read_csv("/Users/Dorene/Desktop/model-dev-coding-exercise-public-master 2/fuelPrices/GDA_TETSTX.csv")

Henry_Hub['Date'] = pd.to_datetime(Henry_Hub['Date'])
GDA_TETSTX['Date'] = pd.to_datetime(GDA_TETSTX['Date'])

Henry_Hub1718 = Henry_Hub[pd.DatetimeIndex(Henry_Hub['Date']).year.isin([2017,2018])]
GDA_TETSTX1718 = GDA_TETSTX[pd.DatetimeIndex(GDA_TETSTX['Date']).year.isin([2017,2018])]

Henry_Hub1718['year'] = pd.DatetimeIndex(Henry_Hub1718['Date']).year
Henry_Hub1718['month'] = pd.DatetimeIndex(Henry_Hub1718['Date']).month
GDA_TETSTX1718['year'] = pd.DatetimeIndex(GDA_TETSTX1718['Date']).year
GDA_TETSTX1718['month'] = pd.DatetimeIndex(GDA_TETSTX1718['Date']).month

Henry_Hub1718J = Henry_Hub1718.merge(data, how='left', left_on=['Variable', 'year','month'], right_on=['FuelPriceName', 'Year','Month'])
GDA_TETSTX1718J = GDA_TETSTX1718.merge(data, how='inner', left_on=['Variable', 'year','month'], right_on=['FuelPriceName', 'Year','Month'])


Joined = pd.concat([Henry_Hub1718J,GDA_TETSTX1718J], axis = 0)

Joined['RunningCost'] = (Joined['Price'] + Joined['FuelTransportationCost'])*Joined['HeatRate'] + Joined['VOM']

Joined['RunningCost']
```
